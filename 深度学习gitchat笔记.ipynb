
# 感知机（一个神经元）
## 一、特点
* 接收多个输入信号，输出一个信号。

## 二、例子
$
y = \left\{\begin{matrix}
1 (w_1x_1 + w_2x_2 \geq \theta ) 
\\
0 (w_1x_1 + w_2x_2 <  \theta ) 
\end{matrix}\right.
$

写法二：
$y = h(b + w_1_x1 + w_2x_2)$


看上面这个例子，$x_1$和$x_2$分别是两个输入信号，最终得到一个输出信号$y$
* $w_1$和$w_2$：两个输入信号的权重
* $\theta$：阈值（大于这个阈值，输出才能是1，才能激活函数），在神经网络世界里，阈值表示神经元被激活的容易程度。

## 三、激活函数
上面的“写法二”中的$h(x)$ 函数会将输入信号的总和转换为输出信号，这种函数，就是**激活函数**（activation function）。
* 作用： 判断输入信号x是否能够让输出信号为1（激活函数内部包含了阈值的比较）
